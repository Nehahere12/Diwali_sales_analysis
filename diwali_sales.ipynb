```python
#total no. of orders from top 10 states


sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(data=sales_state,x='State',y='Orders' ,hue='State')
```

```python
#total no. of amount from top 10 states


amount_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,7)})
sns.barplot(data=amount_state,x='State',y='Amount' ,hue='State')
```

***Uttar pradesh have the highest number of sales and orders***

***marital status***

```python
ax=sns.countplot(data=df,x='Marital_Status' ,hue='Marital_Status',legend=False)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Status_shadi=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(5,7)})
sns.barplot(data=Status_shadi,x='Marital_Status',y='Amount' ,hue='Marital_Status' ,legend=False)
```

***Unmarries buy more orders as compared to married people in terms of sales and orders***

***occupation***

```python
ax=sns.countplot(x='Occupation',data=df,hue='Occupation')
sns.set(rc={'figure.figsize':(25,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Ocs=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(data=Ocs,x='Occupation',y='Amount' ,legend=False,hue='Occupation')
```

***Healthcare has most numbers of orders whileIT sector contributes the most to the amount and the sales***

***Product category***


```python
ax=sns.countplot(x='Product_Category',data=df,hue='Product_Category')
sns.set(rc={'figure.figsize':(30,15)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Ocs=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(30,7)})
sns.barplot(data=Ocs,x='Product_Category',y='Amount' ,legend=False,hue='Product_Category')
```
